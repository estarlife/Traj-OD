```python
def loadData(filename):
    pkl_file = open(filename, 'rb')
    data = pickle.load(pkl_file)

    pkl_file.close()
    #loader = np.load(filename)
    return data
traj=loadData('trajectory')
traj
```

```python
def traj_partition(traj):
    all_endpoint=[]
    endpoint=[]

    
    for i in range(len(traj)):
        tmp=traj[i][0][0]    
        tmp_index=0
        endpoint=[]
        for j in range(len(traj[i])):
            
#           print(j, len(traj[i]))
            if j==len(traj[i])-1:
                endpoint.append([tmp_index,j])
            elif  traj[i][j][0]-tmp>10000:
                endpoint.append([tmp_index,j])
                tmp_index=j+1
                tmp=traj[i][j+1][0]
        all_endpoint.append(endpoint)
    return all_endpoint
        
endpoint=traj_partition(traj)
endpoint
```

```python
def subtrajectory(traj, endpoint):  #endpoint in form [traj_index, []]
    traj_index, [st, ed]=endpoint
    subtraj=traj[traj_index][st:ed+1]
    #list slice：[start，stop)，include start but not stop
    return subtraj

def subtrajectoryfromst2ed(traj, endpoint): #endpoint in form [traj_index, []]
    traj_index, [st, ed]=endpoint
    subtraj=[traj[traj_index][st],traj[traj_index][ed]]
    return subtraj
```
