```python
def loadData(filename):
    pkl_file = open(filename, 'rb')
    data = pickle.load(pkl_file)

    pkl_file.close()
    #loader = np.load(filename)
    return data
traj=loadData('trajectory')
traj
```

```python
trajectory=traj[0]
def plot_trajectory(trajectory):
    plt.plot(trajectory[0:len(trajectory),1], trajectory[0:len(trajectory),2])
```

```python
plot_trajectory(trajectory)
```

```python
def subtrajectory(traj, endpoint): #endpoint in form [traj_index, []]
    traj_index, [st, ed]=endpoint
    subtraj=traj[traj_index][st:ed+1]
    #list slice：[start，stop)，include start but not stop
    return subtraj
plot_trajectory(subtrajectory(traj, [0, [1,5]]))
```
